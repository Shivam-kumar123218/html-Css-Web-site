```python
data['received_by'] = data['received_by'].replace({'Mr.': 'Male', 'Mrs.': 'Female'})
data
```

```python
data['received_by'].value_counts()
```

```python
data = data.rename(columns={'received_by': 'gender' ,'transaction_amount': 'transaction_' })
data
```

```python
data['transaction_'] = '' + data['transaction_'].astype(str)
data
```

```python
sns.countplot(x='item_type' , hue='gender', data=data)
```

```python
sns.barplot(x='gender' , y='transaction_' , data=data)
```

```python
sns.boxplot(x='item_type' , y='quantity' ,hue='gender', data=data)
```

```python
plt.pie(data['gender'].value_counts(), labels=data['gender'].unique(), autopct='%1.1f%%')
plt.title('Item Type Distribution')
plt.show()
```

```python
plt.pie(data['transaction_type'].value_counts(), labels=data['transaction_type'].unique(), autopct='%1.1f%%')
plt.title('Item Type Distribution')
```
