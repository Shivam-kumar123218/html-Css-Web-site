```python
import numpy as np
import pandas as pd
```

```python
data = {
'A': [1,2,np.nan,4,np.nan],
'B': [1,np.nan,3,np.nan,5],
'C': [1,2,3,4,5,],
'D': [ np.nan,2,3,np.nan,5],
}
pd =pd.DataFrame(data)
```
