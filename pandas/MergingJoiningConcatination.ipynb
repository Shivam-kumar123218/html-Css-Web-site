```python
import numpy as np
import pandas as pd
```

```python
#Merging 2 dataframes
```

```python
employees = pd.DataFrame({
       'employee_id': [1,2,3,4,5],
        'Name':['shivam','sujeet','rich','khemraj','vineeta'],
        'Department':['data science','Trader','money','job','house wife']
})

Salary = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
Salary
```

```python
pd.merge(employees,Salary,on='employee_id',how ='inner')
```

```python
pd.merge(employees,Salary,on='employee_id',how='outer')
```

```python
pd.merge(employees,Salary,on='employee_id',how='left')
```

```python
pd.merge(employees,Salary,on='employee_id',how='right')
```

```python
sh = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

sh1 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
sh
```

```python
pd.concat([sh,sh1])
```

```python
#Joining 2 data frames
```

```python
sh3 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
sh2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
sh3
```

```python
sh2
```

```python
sh3.join(sh2,how='outer')
```

```python
sh2.join(sh3)
```
